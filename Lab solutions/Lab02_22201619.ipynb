```python
#You must run this cell to print matrix and for the driver code to work
def print_matrix(m):
  row,col = m.shape
  for i in range(row):
    c = 1
    print('|', end='')
    for j in range(col):
      c += 1
      if(len(str(m[i][j])) == 1):
        print(' ',m[i][j], end = '  |')
        c += 6
      else:
        print(' ',m[i][j], end = ' |')
        c += 6
    print()
    print('-'*(c-col))
```

```python
#task-4
def atm_triangle(n):
  #pass
  arr=np.zeros((n,n),int)

  for i in range(len(arr)):
    if i==0:
      arr[0][0]=1
    elif i==1:
      arr[1][0]=2
      arr[1][1]=2
    else:
      for j in range(i):
        if j==0:
          arr[i][0]=i+1
          arr[i][i]=i+1
        else:
          sum=0
          for k in range(0,j+1):
            sum+=arr[i-1][k]
          arr[i][j]=sum

  return arr

def print_without_none(matrix):
  for i in range(len(matrix)):

    for j in range(len(matrix[i])):
        if matrix[i][j]!=0:
          print(matrix[i][j],end=' ')
    print('')

  #this prints the non None elements of matrix
  pass


n = 5
returned_value = atm_triangle(n)
print_without_none(returned_value)
#This should print
# 1
# 2  2
# 3  4  3
# 4  7  10  4
# 5  11 21  25  5
```
