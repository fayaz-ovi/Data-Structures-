```python
class Patient:

  def __init__(self,id,name,age,b_g,next,priv):

    self.id=id
    self.name=name
    self.age=age
    self.b_g=b_g
    self.next=next
    self.priv=priv


class WRM:

  def __init__(self):

    self.patients=Patient(None,None,None,None,None,None)
    self.patients.next=self.patients
    self.patients.priv=self.patients
    self.temp1=self.patients


  def RegisterPatient(self,id,name,age,b_g):

    pat=Patient(id,name,age,b_g,self.patients,None)

    self.temp1.next=pat
    pat.priv=self.temp1

    self.patients.priv=pat

    self.temp1=pat

    print(f"Successfully added {self.temp1.name} as a patient")

  def ServePatient(self):


    if self.patients.next==self.patients:
      return f"No patients is waiting."

    if self.patients.next.next!=self.patients:
      pat=self.patients.next
      self.patients.next=pat.next
      self.patients.next.priv=self.patients
      return (f"{pat.name} has been served")

    elif self.patients.next.next==self.patients:
      pat=self.patients.next
      self.patients.next=pat.next
      self.patients.next.priv=self.patients
      self.temp1=self.patients
      return (f"{pat.name} has been served")


  def CancelAll(self):

    self.patients.next=self.patients
    self.patients.priv=self.patients
    self.temp1=self.patients
    print('All the appointments has been canceled')

  def CanDoctorGoHome(self):

    if self.patients==self.patients.next:

      return "Yes"
    else:

      return "No"

  def ShowAllPatients(self):

    head=self.patients
    temp3=head.next

    if head!=temp3:

     while temp3!=head:
      print(temp3.name, end=" ")
      temp3=temp3.next
     print('')

    else:
      print("No patient is waiting in the room")


  def ReverseTheLine(self):

    head1=Patient(None,None,None,None,None,None)
    temp1=head1

    head=self.patients
    temp=head.priv

    while temp!=head and temp!=None :
      pat=Patient(temp.id,temp.name,temp.age,temp.b_g,head1,temp1)
      temp1.next=pat
      temp1=pat
      temp=temp.priv

    head1.priv=temp1
    self.temp1=temp1

    self.patients=head1

##########################################################################################
##########################################################################################
##########################################################################################
##########################################################################################



hospital = WRM()
print('Welcome to Waiting Room Management System')
while True:
  print(f'==Choose an Option==\n1. RegisterPatient()\n2. ServePatient()\n3. CancelAll()\n4. CanDoctorGoHome()\n5. ShowAllPatient()\n6.Reverse The Line()\n7. Exit\n====================')

  user = int(input('Enter your choice:'))

  if 1 > user or user > 7:
    print('Invalid number')
  elif user == 1:
    hospital.RegisterPatient(int(input('Enter Your ID:')),input('Enter Your Name:'),int(input('Enter Your Age:')),input('Enter Your bloodgroup:'))
    print('Success registering patient')
  elif user == 2:
    print(hospital.ServePatient())
  elif user == 3:
    hospital.CancelAll()
  elif user == 4:
    print(hospital.CanDoctorGoHome())
  elif user == 5:
    hospital.ShowAllPatients()
  elif user == 6:
    hospital.ReverseTheLine()
  elif user==7:
    print('Thank You for Using waiting Room management System\nEXITING....')
    break



```
