```python
#array based stack with limit
class Stack:
  length_arr=0
  def __init__(self,len,typ= int):
    self.__arr=np.zeros(len, typ) #[0,0,0,0,0,0,,0,0]
    self.__count=0
    Stack.length_arr=len

  def push(self,elem):
    if self.__count<Stack.length_arr:
      self.__arr[self.__count]=elem
      self.__count+=1
    else:
      return "Stack Overflow"

  def peak(self):
    if self.__arr[0]==0:
      return None
    else:
      return self.__arr[self.__count-1]

  def pop(self):
    if self.__arr[0]==0:
      return "Stack underflow"
    else:
      pop_item=self.__arr[self.__count-1]  #[10,20,30,0] s_c=3
      self.__arr[self.__count-1]=0    #[10,20,0,0]
      self.__count-=1
      return pop_item

  def isEmty(self):
    if self.__count==0:
      return "Not empty"
    else:
      return "Empty"
```

```python
#Linked List based stack with limit
class Stack:
  limit=0
  def __init__(self,limit):
    self.__head=Node(None,None,None)
    self.__head.priv=self.__head
    self.__head.next=self.__head
    self.__count=0
    self.__temp=self.__head
    Stack.limit=limit

  def push(self,elem):
    if self.__count<Stack.limit:
      a=Node(elem,self.__head,self.__temp)
      self.__temp.next=a
      self.__temp=a
      self.__head.priv=a
      self.__count+=1

    else:
      return "stack overflow exeption"

  def pop(self):
    if self.__head==self.__temp:
      return "stack underflow"
    else:
      poped=self.__temp
      self.__temp=self.__temp.priv
      self.__temp.next=self.__head
      self.__head.priv=self.__temp
      self.__count-=1
      return poped.elem

  def peak(self):
    if self.__head==self.__temp:
      return "empty stack"
    else:
      return self.__temp.elem

  def isEmty(self):
    if self.__head!=self.__temp:
      return "Not empty"
    else:
      return "Empty"
```

```python
# mid summer-23 Ques-1

def row_max_min(arr):

  max_arr=np.zeros(len(arr),int)

  for i in range(len(arr)):
    max=arr[i][0]
    for j in range(1,len(arr[0])):
      if arr[i][j]>max:
        max=arr[i][j]
    max_arr[i]=max
  min=max_arr[0]
  for i in range(1,len(max_arr)):
    if min>max_arr[i]:
      min=max_arr[i]

  return min

arr1=np.array([[2,7,5],[9,3,8],[1,4,6]])
arr2=np.array([[2,7,5,10],[9,3,8,4]])

print(row_max_min(arr1))
print(row_max_min(arr2))
```
